# Operators

```python
# Arithmatic operator

a= 17
b= 5
print(a+b)
print(a-b)
print(a*b)
print(a/b)
print(a%b)
print(a**b)
```

```python
# Comparsion operator
a=7
b=5
print(a<b)
print(a>b)
print(a>=b)
print(a<=b)
print(a==b)
print(a!=b)
```

```python
# Logical operators
a=True
b=False
print(a and b)
print(a or b)
print(not b)
```

```python
# Bitwise operator
a=5
b=2
print(a&b)
print(a|b)
print(a^b)
print(~a)
print(a>>b)
print(a<<b)
```
